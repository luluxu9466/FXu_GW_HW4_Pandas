```python
# Dependencies
import pandas as pd
```

```python
# load CSV
csvfile = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas
df = pd.read_csv(csvfile, low_memory = False)
```

```python
# Inspect all columns
df.columns
```

```python
# Extract only columns 0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
df = df.iloc[:,[0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]]
df
```

```python
# Change "0" to "No" and "1" to "Yes" in response columns
df = df.replace(0, "No")
df = df.replace(1, "Yes")
df.head()
```

```python
# Calculate total number of respondents in survey
respondent_num = len(df.index)
respondent_num
```

```python
# Extract rows corresponding only to people who attended a bootcamp
# df = df.groupby("AttendedBootcamp")
attended_df = df.loc[df["AttendedBootcamp"] == "Yes", :]
attended_df.head()
```

```python
# Calculate average age of attendees
average_age = attended_df["Age"].dropna().mean()
average_age
# Calculate how many people attended a bootcamp
attended_num = attended_df["AttendedBootcamp"].count()
attended_num

# Calculate how many attendees hold degrees
hold_degrees = attended_df["SchoolDegree"].count()
hold_degrees

# Calculate the percentage of people who attended a bootcamp and hold a college degree
degree_percentage = hold_degrees / len(attended_df.index) *100


# Count number of attendees who self-identify as male; female; or are of non-binary gender identification
total_gender = attended_df["Gender"].count()
male_respondents = attended_df["Gender"].value_counts()["male"]
female_respondents = attended_df["Gender"].value_counts()["female"]
nongender_respondets = total_gender - male_respondents - female_respondents

# Calculate percentage of respondents who attended a bootcamp
percentage_attended = attended_num / respondent_num *100
percentage_attended


# Calculate percentage of respondents belonging to each gender
male_percent = male_respondents/total_gender * 100
male_percent

female_percent = female_respondents/total_gender * 100
female_percent

non_gender_specific_percent = nongender_respondets/total_gender * 100
non_gender_specific_percent

# Calculate percentage of attendees with a school degree
percentage_degree = hold_degrees/attended_num * 100

# Calculate average post-bootcamp salary
average_salary = attended_df["BootcampPostSalary"].dropna().mean()
average_salary
```

```python
# Create a new table consolodating above calculations
bootcamp_table = pd.DataFrame({"Total Surveyed": [respondent_num],
                               "Total Bootcamp attendees": [attended_num],
                               "% attended bootcamp": [percentage_attended],
                               "Average Age": [average_age],
                               "% Male": [male_percent],
                               "% Female": [female_percent],
                               "% Non Gender Specific": [non_gender_specific_percent],
                               "Has a degree":[degree_percentage],
                               "Average Post Bootcamp Salary": [average_salary]
})
bootcamp_table.dtypes
```

```python
# Improve formatting before outputting spreadsheet
bootcamp_table["Average Age"] = bootcamp_table["Average Age"].map("{:,.2f}".format)
bootcamp_table["% Male"] = bootcamp_table["% Male"].astype(float).map("{:,.2f}%".format)
bootcamp_table["% Female"] = bootcamp_table["% Female"].astype(float).map("{:,.2f}%".format)
bootcamp_table["% Non Gender Specific"] = bootcamp_table["% Non Gender Specific"].astype(float).map("{:,.2f}%".format)
bootcamp_table["% attended bootcamp"] = bootcamp_table["% attended bootcamp"].astype(float).map("{:,.2f}%".format)
bootcamp_table["Has a degree"] = bootcamp_table["Has a degree"].map("{:,.2f}%".format)
bootcamp_table["Average Post Bootcamp Salary"] = bootcamp_table["Average Post Bootcamp Salary"].map("${0:,.0f}".format)
bootcamp_table
```

```python
# Export to Excel
bootcamp_table.to_excel("output/Bootcamppart1.xlsx", index=False)
```
